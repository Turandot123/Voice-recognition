```python
from io import BytesIO
import tempfile
import requests
import numpy as np
import scipy.signal as sg
import pydub
import matplotlib.pyplot as plt
from IPython.display import Audio, display
from pydub import AudioSegment
import array
%matplotlib inline
```

```python
audio = AudioSegment.from_mp3('../input/Voice_Elena.mp3')

sample=audio.get_array_of_samples()

```

```python
b,a =sg.butter(3,1/200, btype='low')
```

```python
# pass a low filter to the audio file already converted to array.array
# convert low to array 
low= sg.filtfilt(b, a, sample)
low =list(map(int, list(low)))
low_pass=array.array('h')
low_pass.fromlist(low)
```

```python
low_pass
```

```python
# convert array to audio file recognised by pydub

new_sound = audio._spawn(low_pass)
```

```python
# play new sound
new_sound
```

```python
# print original audio

plt.plot(sample)
```

```python
# print new audio with low-pass filter

plt.plot(low_pass)
```

```python
b,a =sg.butter(3,1/400, btype='high')
high= sg.filtfilt(b, a, sample)
high =list(map(int, list(low)))
high_pass=array.array('h')
high_pass.fromlist(high)
```

```python
high_pass
```

```python
new_sound = audio._spawn(high_pass)
new_sound
```
